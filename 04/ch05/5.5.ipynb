# 方策オフ型と重点サンプリング

```python
# 初めに確率分布$\pi$の期待値を、通常のモンテカルロ法を使って求める。
import numpy as np

x = np.array([1, 2, 3])
pi = np.array([0.1, 0.1, 0.8])

# 期待値
e_pi = np.sum(x * pi)
print("E_pi(x)", e_pi)

# モンテカルロ法
n = 100
samples = []
for _ in range(n):
    s = np.random.choice(x, p=pi)   # piを使ってサンプリング
    samples.append(s)

mean = np.mean(samples)
var = np.var(samples)
print("Monte Carlo: {:.2f} (var: {:.2f})".format(mean, var))
```

```python
# 重点サンプリングを用いて期待値を求める

b = np.array([1/3, 1/3, 1/3])
n = 100
samples = []

for _ in range(n):
    idx = np.arange(len(b)) # [0, 1, 2]
    i = np.random.choice(idx, p=b)  # bを使ってサンプリング（添え字をサンプリングし、対応するxをサンプリングしていることになっている）。
    s = x[i]
    rho = pi[i] / b[i]
    samples.append(rho * s)

mean = np.mean(samples)
var = np.var(samples)
print("IS: {:.2f} (var: {:.2f})".format(mean, var)) # 確率分布bは、どの確率変数をとる確率も等しいことを指し示しているので分散（ばらつき）が大きくなる。
```

ばらつきが大きくなるのは、重み付きのデータそのものにばらつきがあるためである。言い換えれば、サンプリングされる値が極端であるため分散が大きくなるのだ。

$1$が選ばれたとき、その重み付きデータは$1 \times 0.1 \div (1 / 3) = 0.3$、$2$のときは$2 \times 0.1 \div (1 / 3) = 0.6$、$3$のときは$3 \times 0.8 \div (1 / 3) = 7.2$であり、重みを加味した際の確率変数は`[0.3, 0.6, 7.2]`となり、これが等確率でサンプリングされるので、その分散は大きくなる（見れば分かる通り、7.2以外は1未満であり、数値が二極化している。

## 5.5.3 分散を小さくするには

上述したことから、分散が大きくなるのは重み$\rho$が掛かっているためであると考えられる。そこでこの$\rho = \pi / b$が$1$に十分小さい、つまり$\pi \risingdotseq b$であれば$\rho$を$1$に十分近づけることができる。

```python
b = np.array([0.2, 0.2, 0.6])   # 確率分布を変更
n = 100
sample = []

for _ in range(n):
    idx = np.arange(len(b)) # [0, 1, 2]
    i = np.random.choice(idx, p=b)
    s = x[i]
    rho = pi[i] / b[i]
    samples.append(rho * s)

mean = np.mean(samples)
var = np.var(samples)
print("IS: {:.2f} (var: {:.2f})".format(mean, var))
```
