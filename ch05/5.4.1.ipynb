# 単純なレイヤの実装

リンゴの買い物の例を、Pythonで実装する。ここでは、計算グラフの乗算ノードを「乗算レイヤ（MulLayer; Multiple Layer）」、加算ノードを「加算レイヤ（AddLayer; Addition Layer）」という名前で実装する。

二つのレイヤは`layer_naive.py`に保存する。

## 乗算レイヤの実装

```python
# 乗算レイヤを用いて、リンゴの問題（問1）を実装する。

from layer_naive import MulLayer

apple = 100
apple_num = 2
tax = 1.1

# layer
mul_apple_layer = MulLayer()
mul_tax_layer = MulLayer()

# forward
apple_price = mul_apple_layer.forward(apple, apple_num) # まずはリンゴの個数とリンゴ一つの値段を乗算し、リンゴの税抜き価格を計算する
price = mul_tax_layer.forward(apple_price, tax) # リンゴの税抜き価格に消費税を乗算して、合計金額を計算する。

print(price)    # 220.000000003
```

```python
# 各変数に関する微分は、`backward()`メンバメソッドで求めることができる。

# backward
d_price = 1
d_apple_price, d_tax = mul_tax_layer.backward(d_price)

d_apple, d_apple_num = mul_apple_layer.backward(d_apple_price)

print(d_apple, d_apple_num, d_tax)  # 2.2, 110, 200
```
