## 加算レイヤの実装

```python
# ではリンゴとミカンの合計を求める問題（問2）を実装してみましょう。
# まずは自分でやってみましょう。

from layer_naive import MulLayer, AddLayer

apple = 100
apple_num = 2

orange = 150
orange_num = 3

tax = 1.1

# apple_price = MulLayer()
# orange_price = MulLayer()

mul_apple_layer = MulLayer()
mul_orange_layer = MulLayer()
mul_tax_layer = MulLayer()

add_layer = AddLayer()

# リンゴの合計（税抜き）
apple_price = mul_apple_layer.forward(apple, apple_num)

# ミカンの合計（税抜き）
orange_price = mul_orange_layer.forward(orange, orange_num)

# リンゴとミカンの合計（税抜き）
apple_and_orange_price = add_layer.forward(apple_price, orange_price)

# リンゴとミカンの合計（税込み）
price = mul_tax_layer.forward(apple_and_orange_price, tax)

print("forward: " + str(price))

# backward

d_price = 1
d_apple_and_orange_price, d_tax = mul_tax_layer.backward(d_price)
d_apple_price, d_orange_price = add_layer.backward(d_apple_and_orange_price)
d_apple, d_apple_num = mul_apple_layer.backward(d_apple_price)
d_orange, d_orange_num = mul_orange_layer.backward(d_orange_price)

print(f"""
backward: \n
    d_price: {d_price} \n
    d_apple_and_orange_price: {d_apple_and_orange_price} \n
    d_tax: {d_tax} \n
    d_apple_price: {d_apple_price} \n
    d_orange_price: {d_orange_price} \n
    d_apple: {d_apple} \n
    d_apple_num: {d_apple_num} \n
    d_orange: {d_orange} \n
    d_orange_num: {d_orange_num}
""")
```
