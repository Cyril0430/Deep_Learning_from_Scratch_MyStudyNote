```python
import numpy as np
```

```python
# 高さ28、幅28でチャンネル数が1の画像データが10個ある、ということをデータで表現するには4次元配列（形状は`(10, 1, 28, 28)`

x = np.random.rand(10, 1, 28, 28)
x.shape
```

```python
print(x)
print(f"x mean is {np.average(x)}, standard deviation is {np.std(x)}")
```

```python
# 余談（飽きたので少し遊ぶ）
# xの各データを、全ての要素について標準化する

x_mean = np.average(x)
# tmp_x = 1
# for i_x_shape in x.shape:
#     tmp_x *= i_x_shape

# x_mean = np.sum(x) / tmp_x
x_std = np.std(x)

z_x = (x - x_mean) / x_std
print(f"z_x is {z_x}")
print(f"\n z_x\' mean is {np.average(z_x)}, std is {np.std(z_x)}")
```
