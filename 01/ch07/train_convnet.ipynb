```python
import sys, os
sys.path.append(os.pardir)

import numpy as np
from dataset.mnist import load_mnist
from simple_conv_net import *
```

```python
(x_train, t_train), (x_test, t_test) = \
    load_mnist(flatten=False)

x_train, t_train = x_train[:5000], t_train[:5000]
```

```python
train_size = x_train.shape[0]
train_loss_list = []

iters_num = 10000
batch_size = 100
learning_rate = 0.1

network = \
    SimpleConvNet(input_dim=(1,28,28), conv_param = {'filter_num': 30, 'filter_size': 5, 'pad': 0, 'stride': 1}, hidden_size=100, output_size=10, weight_init_std=0.01)
```

```python
# 多分、自分で作ったPoolingレイヤよりも、斎藤氏のもののほうが処理速度が速いと思うので、そっちを使うようにしましょう。
for i in range(iters_num):
    batch_mask = np.random.choice(train_size, batch_size)
    x_batch = x_train[batch_mask]
    t_batch = t_train[batch_mask]

    grad = network.gradient(x_batch, t_batch)

    for key in ('W1', 'b1', 'W2', 'b2', 'W3', 'b3'):
        network.params[key] -= learning_rate * grad[key]
    
    loss = network.loss(x_batch, t_batch)
    train_loss_list.append(loss)
```
