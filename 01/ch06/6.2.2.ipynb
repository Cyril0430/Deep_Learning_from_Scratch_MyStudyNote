# 隠れ層のアクティベーション分布

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.random.randn(1000, 100)  # 1000個のデータ
node_num = 100  # 隠れ層のノードの数
hidden_layer_size = 5   # 隠れ層が5層
activations = {}
```

```python
for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i-1]
    
    w = np.random.randn(node_num, node_num) * 1

    z = np.dot(x, w)
    a = sigmoid(z)
    activations[i] = a
```

```python
# ヒストグラムを描画
for i, a in activations.items():
    plt.subplot(1, len(activations), i + 1)
    plt.title(str(i+1) + "-layer")
    plt.hist(a.flatten(), 30, range=(0, 1))

plt.show()
```

```python
for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i-1]
    
    w = np.random.randn(node_num, node_num) * 0.01  # 標準偏差を1から0.01に変更した

    z = np.dot(x, w)
    a = sigmoid(z)
    activations[i] = a
```

```python
# ヒストグラムを描画
for i, a in activations.items():
    plt.subplot(1, len(activations), i + 1)
    plt.title(str(i+1) + "-layer")
    plt.hist(a.flatten(), 30, range=(0, 1))

plt.show()
```

## Heの初期値

```python
import numpy as np
import matplotlib.pyplot as plt

# def sigmoid(x):
    # return 1 / (1 + np.exp(-x))
# def tanh(x):
#     # return (1 - np.exp(-x)) / (1 + np.exp(-x))
#     return np.tanh(x)
#     # return np.sin(x)**2
#     # return np.max(x, 0)
#     # return np.log(x)
def function_activation(x):
    return np.maximum(0, x)


x = np.random.randn(1000, 100)  # 1000個のデータ
node_num = 100  # 隠れ層のノードの数
hidden_layer_size = 5   # 隠れ層が5層
activations = {}

for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i-1]
    
    # w = np.random.randn(node_num, node_num) * 0.01  # 標準偏差を1から0.01に変更した
    node_num = 100
    w = np.random.randn(node_num, node_num) / np.sqrt(node_num) * np.sqrt(2)

    z = np.dot(x, w)
    # a = sigmoid(z)
    # a = tanh(z)
    a = function_activation(z)
    activations[i] = a

for i, a in activations.items():
    plt.subplot(1, len(activations), i + 1)
    plt.title(f"{i+1}-layer")
    plt.hist(a.flatten(), 30, range = (0, 1))
```
