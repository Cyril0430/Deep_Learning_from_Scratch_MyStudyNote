````python
import numpy as np
````

````python
def step_function_i(x):
    if x > 0:
        a = 1
    else:
        a = 0
    return a

step_function_i(-1)
````

````python
def step_function_ii(x):
    y = x > 0
    return y.astype(int)
````

````python
x = np.array([-1.0, 1.0, 2.0])
step_function_ii(x)
````

````python
x = np.array([-1.0, 1.0, 2.0])
x
````
