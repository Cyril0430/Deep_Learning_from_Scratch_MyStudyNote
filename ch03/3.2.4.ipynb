```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([1.0, 2.0, 3.0])
sigmoid(x)
```

```python
t = np.array([1.0, 2.0, 3.0])

1.0 / t
```

```python
import matplotlib.pylab as plt
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show
```
