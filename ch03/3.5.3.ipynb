# ソフトマックス関数の特徴

```python
import numpy as np

def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    y = exp_a / np.sum(exp_a)
    return y

a = np.array([0.3, 2.9, 4.0])
y = softmax(a)
print(y)
print(np.sum(y))
```

任意の配列に対し、そのソフトマックス関数の出力値の合計は${1}$になる。

この性質によって、ソフトマックス関数は確立として解釈することができる。

ソフトマックス関数は単調増加関数であるため、出力する前の数値の大小関係がそのまま出力の大小関係となる。そのため、ソフトマックス関数による出力は省略されることがしばしばある。
