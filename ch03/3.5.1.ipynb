ソフトマックス関数
$${
  y_k = \frac{\exp(a_k)}{\sum_{i = 1}^n \exp(a_i)}
}$$

を実装する。ただしこれは出力層のノード ${y_{k\in[1, n]}}$ に対するその一つ前の隠れ層からの入力 ${a_{k\in[1, n]}}$ を示している。

```python
import numpy as np

a = np.array([0.3, 2.9, 4.0])
```

```python
exp_a = np.exp(a)
```

```python
print(exp_a)
```

```python
sum_exp_a = np.sum(exp_a)   # exp_aの各要素の総和
print(sum_exp_a)
```

```python
# ソフトマックス関数の定義

def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y

a = np.array([0.3, 2.9, 4.0])
softmax(a)
```
